Welcome to Matplotlib Library

```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Draw Line Graph

x=[1,2,3,4]
x2=[1,1.5,2,4]
y=[10,20,30,40]
plt.plot(x,y,marker='x',c='black')
#To show the graph
plt.show()
```

```python
#Add label and legend
plt.plot(x,y,label='line 1',marker = 'x', c="blue")
plt.plot(x2,y,label='line 2',marker = 'o', c="red")
plt.legend(loc=0)
plt.title("Line plot") 
plt.xlabel("x-label") 
plt.ylabel("y-label")
plt.show()
```

```python
#Draw a Scatter plot
x=[1,2,3,4,5,6,7,8,9]
x2 = [1,3,3,4,2,6,7.5,8,9]
y=[2,1,4,6,2,7,8,4,8]
plt.scatter(x,y,c='red',label='points 1')
plt.scatter(x2,y,c='b',label='points 2')
plt.legend()
plt.show()
```

```python
data=np.random.normal(3,1,1000)
plt.hist(data)
plt.title("this is a histogram graph") 
plt.xlabel("x-label") 
plt.ylabel("y-label")
```
