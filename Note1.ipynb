```python
#to combine two value into new attribute
df1['age_weight']= df1.age.astype(str) + ',' + df1.weight.astype(str)
```

```python
df1.head()
```

```python
#Drop an attribute (column) to remove the attribute use this code
df1=df1.drop('age_weight', axis=1)
df1
```

```python
#to sort the value in ascending or descending order use this code
df1=df1.sort_values('age',ascending=True)
df1['age']
```

```python
# to sort the value if the boolean is false the data will sort into descending order
df1=df1.sort_values('age',ascending=False)
df1['age']
```

```python
#to sort the value on the selected attribute use this code
df1.sort_values('weight')
```
